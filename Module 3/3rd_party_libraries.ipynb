```python
from datetime import datetime

import dateutil
import pytz
```

```python
def convert(source_file, target_file):
    tz_eastern = pytz.timezone('US/Eastern')

    with open(source_file) as source:
        with open(target_file) as target:
            csv_reader = csv.reader(source)
            csv_writer = csv.writer(target)

            header = next(csv_reader)
            csv_writer.writerow(header)

            for dt, rate in csv_reader:
                dt_naive = parser.parse(dt).replace(hour=17)
                dt_aware = dt_nyc.localize(dt_naive)
                dt_utc = dt_aware.astimezone(pytz.UTC)
                csv_writer.writerow([dt_utc.isoformat(), rate])
```

```python
def convert(source_file, target_file):
    tz_eastern = pytz.timezone('US/Eastern')

    with open(source_file) as source:
        with open(target_file, 'w') as target:
            csv_reader = csv.reader(source)
            csv_writer = csv.writer(target)

            header = next(csv_reader)
            csv_writer.writerow(header)

            for dt, rate in csv_reader:
                dt_naive = parser.parse(dt).replace(hour=17)
                dt_aware = dt_nyc.localize(dt_naive)
                dt_utc = dt_aware.astimezone(pytz.UTC)
                csv_writer.writerow([dt_utc.isoformat(), rate])
```

```python
from decimal import Decimal
from datetime import date
```

```python
d = {
    "symbol": "IBM",
    "date": date(2020, 9, 21),
    "day": {
        "open": Decimal('120.48'),
        "high": Decimal('120.70'),
        "low": Decimal('118.58'),
        "close": Decimal('120.25'),
        "volume": 5_205_413
    }
}
```

```python
def stock_encoder(obj):
    if isinstance(obj, date):
        return obj.isoformat()
    if isinstance(obj, Decimal):
        return str(obj)
    raise TabError
```

```python
print(json.dumps(d, default=stock_encoder, indent=2))
```

```python
def stock_encoder(obj):
    if isinstance(obj, date):
        return obj.isoformat()
    if isinstance(obj, Decimal):
        return round(float(obj), 2)
    raise TabError
```

```python
print(json.dumps(d, default=stock_encoder, indent=2))
```
