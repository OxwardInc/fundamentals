```python
class Circle:
    def __init__(self, center_x, center_y, radius):
        self.center = center_x, center_y
        self.radius = radius

    def area(self):
        return pi * self.radius ** 2

    def translate(self, x, y):
        self.center = (self.center[0] + x, self.center[1] + y)

    def scale(self, factor):
        self.radius *= factor
```
